## Plots of damping coefficient

```python
mu_open = np.full(500, 0.5)

mu_palm = np.full(500, 0.5)
mu_palm[-50:] = 200

mu_har2 = np.full(500, 0.5)
mu_har2[249:251] = 5000

mu_har3 = np.full(500, 0.5)
mu_har3[166:168] = 5000
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(mu_palm)

ax.set_xticks([0, 500], ["0", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$\mu$")

plt.savefig("mu_palm.png", dpi=100)
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(mu_har2)

ax.set_xticks([0, 250, 500], ["0", "L/2", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$\mu$")

plt.savefig("mu_har2.png", dpi=100)
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(mu_har3)

ax.set_xticks([0, 167, 500], ["0", "L/3", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$\mu$")

plt.savefig("mu_har3.png", dpi=100)
```

```python
p0_p = np.zeros(500)
p0_n = np.zeros(500)
p0_h = np.zeros(500)

p0_p[0:450]=[0.03*i/450 for i in range(0,450)]
p0_p[450:499]=[0.03*(49-i)/49 for i in range(0,49)]

p0_n[0:300]=[0.03*i/300 for i in range(0,300)]
p0_n[300:499]=[0.03*(199-i)/199 for i in range(0,199)]

p0_h[445:455] = [3, 6, 9, 12, 15, 15, 12, 9, 6, 3]
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(p0_p)

ax.set_xticks([0, 500], ["0", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$u(x,0)$")

plt.savefig("p0_p.png", dpi=100)
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(p0_n)

ax.set_xticks([0, 500], ["0", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$u(x,0)$")

plt.savefig("p0_n.png", dpi=100)
```

```python
f, ax = plt.subplots(layout="constrained", figsize=(5, 3))

ax.plot(p0_h)

ax.set_xticks([0, 500], ["0", "L"])
ax.set_yticks([])
ax.set_xlabel("$x$")
ax.set_ylabel("$\\frac{\partial u}{\partial t}(x,0)$")

plt.savefig("p0_h.png", dpi=100)
```
